# Line Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)

plt.show()
```

```python
import numpy as np
x=[1,2,3,4]
y=[20,10,50,30]
plt.plot(x,y)
plt.xticks(np.arange(1,6,step=1))
plt.yticks(np.arange(10,60,step=10))
plt.show()
```

```python
import numpy as np
x=[1,2,3,4]
y=[20,10,50,30]
plt.plot(x,y,color='r',marker = 'o',ls='--')
plt.xticks(np.arange(1,6,step=1))
plt.yticks(np.arange(10,60,step=10))
plt.title('LINE PLOT',color='r',fontsize=25)
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[10,20,40,50,30]
y2=[60,10,20,20,10]
y3=[30,30,10,10,50]
y4=[40,60,50,40,40]

plt.plot(x,y1,marker='o',label='Royal_Enfiled')
plt.plot(x,y2,marker='o',label='Himalyan')
plt.plot(x,y3,marker='o',label='Harley')
plt.plot(x,y4,marker='o',label='Ninja')

plt.title('Bike Details Using Plot')
plt.xlabel('no. of days')
plt.ylabel('Distance covered in KMs')
plt.yticks(np.arange(10,70,step=5))

plt.legend()
plt.grid()
plt.show()
```

# Bar Plot

```python
x=[1,2,3,4,5]
y1=[10,20,40,50,30]

plt.bar(x,y1,label='Royal_Enfiled')
plt.title('Bike Details Using Plot')
plt.xlabel('no. of days')
plt.ylabel('Distance covered in KMs')

plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[10,20,40,50,30]

plt.barh(x,y1,label='Royal_Enfiled')
plt.title('Bike Details Using Plot')
plt.xlabel('no. of days')
plt.ylabel('Distance covered in KMs')

plt.legend()
plt.show()
```

# Scatter Plot

```python
x=[1,2,3,4,5]
y1=[10,20,40,50,30]

plt.scatter(x,y1,label='Royal_Enfiled')
plt.title('Bike Details Using Plot')
plt.xlabel('no. of days')
plt.ylabel('Distance covered in KMs')

plt.legend()
plt.show()
```

# Pie Plot

```python
products = ["I Phone","Laptop","Airpods","Headphone","Keyboard"]
x=[10,20,40,50,30]

plt.pie(x,labels=products)
plt.title('Bike Details Using Plot')
plt.show()
```

# 3D Plot

```python
fig = plt.figure(figsize=(12,6))
fig.add_subplot(projection='3d')
x=[1,2,3,4,5]
y1=[10,20,40,50,30]
y2=[60,10,20,20,10]
y3=[30,30,10,10,50]
y4=[40,60,50,40,40]

plt.plot(x,y1,marker='o',label='Royal_Enfiled')
plt.plot(x,y2,marker='o',label='Himalyan')
plt.plot(x,y3,marker='o',label='Harley')
plt.plot(x,y4,marker='o',label='Ninja')

plt.title('Bike Details Using Plot')
plt.xlabel('no. of days')
plt.ylabel('Distance covered in KMs')

plt.legend()
plt.grid()
plt.show()
```
